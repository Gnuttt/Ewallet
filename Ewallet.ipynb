```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('Transactions.csv',thousands=',')
df1
```

```python
df1['Purchase_status'] = df1['Purchase_status'].fillna('Tự Dùng')
df1
```

```python
df1['Partner_id'] = df1['Partner_id'].fillna(0)
df1
```

```python
df1['Date']= pd.to_datetime(df1['Date'])
df1.dtypes
```

```python
df1['Merchant_per'] = df1['Merchant_id']
df1
```

```python
df1 = df1.replace({'Merchant_per':{22:2, 33:3, 44:4, 55:4, 66:4}})
df1
```

```python
df1['Revenue Per Order'] = df1['Amount']*df1['Merchant_per']/100
df1
```

```python
df1['month'] = pd.DatetimeIndex(df1['Date']).month
df1
```

```python
print(set(df1['month']))
```

```python
REV = df1.groupby('month').sum()['Revenue Per Order']
REV
```

```python
REV.to_csv('REV by Month.csv')
```

```python
duplicateRows = df1[df1.duplicated(['user_id', 'month'])]
```

```python
df1.groupby(df1['user_id'].tolist(),as_index=False).size()
```

```python
dups_id_by_month = df1.pivot_table(columns=['month','user_id'], aggfunc='size')
print (dups_id_by_month)
```

```python
df1.to_csv('transactions_df_cleaned.csv')
```
